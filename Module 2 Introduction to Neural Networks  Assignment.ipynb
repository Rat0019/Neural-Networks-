## 1. Activation Functions:

### Sigmoid Function:

```python
import math

def sigmoid(x):
    return 1 / (1 + math.exp(-x))
```

### Tanh Function:

```python
def tanh(x):
    return (math.exp(x) - math.exp(-x)) / (math.exp(x) + math.exp(-x))
```

### ReLU Function:

```python
def relu(x):
    return max(0, x)
```

# 2. Sample Data:

```python
x1 = 3
x2 = 2
y_actual = 17
```

# 3. Feed Forward:

### Initialize Random Weights:

```python
import random

w1 = random.uniform(1, 10)
w2 = random.uniform(1, 10)
```

### Predict Output (Feed Forward):

```python
y_pred = x1 * w1 + x2 * w2
```

# 4. Error Calculation:

```python
error = (y_actual - y_pred) ** 2
```

# 5. Back Propagation:

### Compute Gradients:

```python
# Partial derivatives of the error with respect to weights
grad_w1 = -2 * x1 * (y_actual - y_pred)
grad_w2 = -2 * x2 * (y_actual - y_pred)
```

### Update Weights Using Learning Rate:

```python
learning_rate = 0.01

# Update weights
w1 = w1 - learning_rate * grad_w1
w2 = w2 - learning_rate * grad_w2
```

# 6. Training:

```python
epochs = 20

for epoch in range(epochs):
    # Feed Forward
    y_pred = x1 * w1 + x2 * w2
    
    # Error Calculation
    error = (y_actual - y_pred) ** 2
    
    # Back Propagation
    grad_w1 = -2 * x1 * (y_actual - y_pred)
    grad_w2 = -2 * x2 * (y_actual - y_pred)
    
    # Update Weights
    w1 = w1 - learning_rate * grad_w1
    w2 = w2 - learning_rate * grad_w2
    
    # Print progress
    print(f"Epoch {epoch + 1}, Error: {error}, y_pred: {y_pred}")
```

# 7. Visualization:

```python
import matplotlib.pyplot as plt

# Initialize lists to store values for plotting
error_progression = []
y_pred_progression = []

for epoch in range(epochs):
    # ... (same code as above)
    
    # Append values for plotting
    error_progression.append(error)
    y_pred_progression.append(y_pred)

# Plotting
plt.figure(figsize=(12, 6))

# Plot Error Progression
plt.subplot(1, 2, 1)
plt.plot(range(1, epochs + 1), error_progression, marker='o')
plt.title('Error Progression')
plt.xlabel('Epochs')
plt.ylabel('Error')

# Plot y_pred Progression
plt.subplot(1, 2, 2)
plt.plot(range(1, epochs + 1), y_pred_progression, marker='o')
plt.title('y_pred Progression')
plt.xlabel('Epochs')
plt.ylabel('y_pred')

plt.tight_layout()
plt.show()
```
